## 03
- Resoluci√≥n de dos sistemas de ecuaciones
    - $ A : \begin{cases} -9x +  4y = 20 \\ -7y + 16x = 80                         \end{cases}$
    
    - $ B : \begin{cases} x - 2y + 3z = 7 \\ 2x + y + z = 4 \\ -3x + 2y -2z = -10 \end{cases}$
- Verifica por c√≥digo los resultados

```python
# Your solution HERE !!!!
import math
import numpy as np
import numpy.linalg as linalg

a = np.array([[-9, 4], [16, -7]])
b = np.array([20, 80])

print(a)
print(b)

x, y = linalg.solve(a, b)
print(f'x={x} y={y}')

print(math.isclose(-9*x + 4*y, 20))
print(math.isclose(16*x - 7*y, 80))

A2 = np.array([[1, -2, 3],
               [2, 1, 1],
               [-3, 2, -2]])
b2 = np.array([7, 4, -10])

x, y, z = linalg.solve(A2, b2)
print(f'x={x} y={y} z={z}')

print(math.isclose(1*x - 2*y + 3*z, 7))
print(math.isclose(2*x + 1*y + z, 4))
print(math.isclose(-3*x + 2*y - 2*z, -10))
```

## 06
- Dada una lista de elementos, crea una funci√≥n que devuelva un dataframe sin los elementos duplicados

```python
def remove_duplicates(lista):
    resultado = []
    for x in lista:
        if x not in resultado:
            resultado.append(x)
    return resultado

serie = ['a','b','c','a','c','a','g']

expected_output = ['a','b','c','g']
output = remove_duplicates(serie)

print("Output:", output)
print("Expected:", expected_output)

assert output == expected_output
```

```python
# contar numeros pares
def contar_pares(lista):
    return sum(1 for x in lista if x % 2 == 0)

print(contar_pares([1,2,3,4,6]))
```

```python
# Sumar numeros impares
def suma_impares(lista):
    return sum(x for x in lista if x % 2 != 0)

print(suma_impares([1,2,3,4,5]))
```

```python
# mayor y menor 
def mayor_menor(lista):
    return max(lista), min(lista)

print(mayor_menor([5,1,9,3]))
```

```python
# numero primos
def es_primo(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(es_primo(7))
```

```python
# promedio
def promedio(lista):
    return sum(lista) / len(lista)

print(promedio([2,4,6]))
```

```python
# Contar caracteres
def contar_caracteres(s):
    return {c: s.count(c) for c in s}

print(contar_caracteres("holaa"))
```

```python
# mayor
def mayor(a,b):
    return a if a > b else b

print(mayor(10,7))
```

```python
#Valores positivos
def positivos(lista):
    return [x for x in lista if x > 0]

print(positivos([-2,3,-1,5]))
```

```python
# Posiciones pares
def posiciones_pares(lista):
    return lista[::2]

print(posiciones_pares([1,2,3,4,5]))
```

```python
# Eliminar negativos
def sin_negativos(lista):
    return [x for x in lista if x >= 0]

print(sin_negativos([-3,4,-1,6]))
```

```python
# Invertir lista
def invertir(lista):
    return lista[::-1]

print(invertir([1,2,3]))
```

```python
# Contar palabras
def contar_palabras(frase):
    return len(frase.split())

print(contar_palabras("Hola mundo Python"))
```

```python
#Quitar espacios
def sin_espacios(s):
    return s.replace(" ", "")

print(sin_espacios("hola mundo"))
```

```python
# Palabra m√°s larga
def palabra_larga(frase):
    return max(frase.split(), key=len)

print(palabra_larga("soy estudiante de python"))
```
