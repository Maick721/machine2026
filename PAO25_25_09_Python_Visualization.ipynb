**2 .plot()**

Múltiples parámetros

plt.plot(x, y, style, label, linewidth=1, alpha=1)

• ‘style’: cadena de texto con formato (color, tipo…) [  color ][ marker ][ line ]

• ‘label’: referencia para leyenda

• ‘linewidth’: brocha

• ‘alpha’: transparencia

Otros: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
y_data = np.random.randint(100,size=(10))
x_data = np.arange(10)
fig,axis = plt.subplots(1,1)
axis.plot(x_data,y_data,'gh--',label='dollars',linewidth=1,alpha=0.6) #creación del gráfico
axis.legend(loc='best') # representar la leyenda
plt.show()
```

```python
ls_count = (284487, 244560, 208493, 196764)
cities = ('Madrid', 'Barcelona', 'Sevilla', 'Valencia')
data = np.array(ls_count)/1000
y_pos = np.arange(data.size)
width = 0.35 # the width of the bars
opacity = 0.7
fig, ax = plt.subplots()
rects1 = ax.barh(y_pos, data, width, alpha=opacity)
rects1[3].set_color('#872f29')
rects1[2].set_color('#ffcb59')
rects1[1].set_color('#d6ee39')
rects1[0].set_color('#6ade30')
ax.set_yticks(y_pos)
ax.set_yticklabels(cities)
# add some text for labels, title and axes ticks
ax.set_ylabel('Ciudades')
ax.set_xlabel('Km de carril bici')
ax.legend((rects1[0], rects1[1], rects1[2], rects1[3]), cities)
plt.show()
```

**6 frame.plot()**

frame.plot(x,y,label=<Series>,ax,style,kind,xticks,yticks,title,subplots)
* (solo para Series) label: referencia para la leyenda
* ax: subfigura en la que dibujar los datos
* style: estilo de la línea
* kind: tipo de gráfico (’bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
* xticks y yticks: valores en los ejes X e Y
* title: string como título
* (solo para DataFrame) subplots: bool si se desean subfiguras separadas para cada columna.
* (solo para DataFrame) ‘x’ e ‘y’ se pueden utilizar para seleccionar columnas para cada eje

```python
#scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')
# preparacion de los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]
for i in groups:
    # busqueda por clase definida
    data.append(iris[iris["species"] == i][columns].values)
#creacion de la grafica
fig = plt.figure()
ax = fig.add_subplot()
for count, d in enumerate(data):
    # separar las variables del grafico
    x, y = d[:,0], d[:,1]
    ax.scatter(x, y, alpha = 0.4, c = colors[count -1], edgecolors = 'none', s= 30, label = groups[count - 1])
    #etiquetas
ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

## 8.1 Matplotlib

### Ejercicio 1: Crear una lista de años y datos aleatorios

1. Crear una lista de años de 2000 a 2020 (eje x).
2. Generar datos aleatorios para el eje y (hacerlo 4 veces distintas: y0, y1, y2 e y3).
3. Representar las 4 secuencias de datos aleatorios en una sola figura utilizando matplotlib.
4. Añadir una leyenda para poder identificar cada secuencia en el gráfico.
5. Nombrar las secuencias de la siguiente forma: “Hawaii”, “San Marino”, “Islas Feroe”, “Guayana”.
6. Representar los mismos datos en 4 subfiguras (axes) distintas, parte de la misma figura.
7. Añadir la siguiente información:
   - Título: “Datos aleatorios”
   - Leyenda
   - Nombre del eje x: “Año”
   - Nombre del eje y: “GDP”
8. Añadir al menos una anotación (texto y flecha) a uno de los gráficos.


```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
years = np.arange(2000, 2021)
np.random.seed(0)

names = ['Hawaii', 'San Marino', 'Islas Feroe', 'Guayana']
data = np.random.normal([100, 120, 80, 90], 5, (len(years), 4))

# 1 Gráfico
plt.figure(figsize=(10, 5))

for i, name in enumerate(names):
    plt.plot(years, data[:, i], label=name)

plt.title('Datos aleatorios')
plt.xlabel('Año')
plt.ylabel('GDP')
plt.legend()

plt.annotate(
    'Punto interesante',
    xy=(2010, data[10, 1]),
    xytext=(2003, 130),
    arrowprops=dict(arrowstyle='->')
)

plt.show()

# Subgráfico
fig, axes = plt.subplots(2, 2, figsize=(10, 7))

for ax, i in zip(axes.flat, range(4)):
    ax.plot(years, data[:, i], label=names[i])
    ax.set_title(names[i])
    ax.set_xlabel('Año')
    ax.set_ylabel('GDP')
    ax.legend()

fig.suptitle('Datos aleatorios')
plt.tight_layout()
plt.show()
```

## 8.2 Seaborn

### Ejercicio 2: Cargar datos y representar relaciones

1. Cargar los datos de un dataset de seaborn (distinto a ‘tips’, ‘iris’ o ‘flights’).
   - Puedes usar el repositorio de datos de seaborn.
   - Asegúrate de utilizar la versión ‘Raw’ si usas la URL en el método read_csv.
   
2. Representar visualmente la relación de cada pareja de atributos (variables).
3. Escoger una pareja (ambos valores continuos y numéricos) y demostrar su correlación.
4. Representar los valores de dos variables (x e y, numéricas) en base a otra variable (categórica).
5. Crear los gráficos que demuestren lo siguiente en el dataset seleccionado:
   - La distribución de valores (media, distribución, outliers) de una variable numérica continua.
   - Comparar la distribución de esa variable con otra (continua y numérica) para representar cuantitativamente la densidad de elementos por valor en ambas variables.
   - Mostrar un ejemplo en el dataset de una variable a la que un modelo de regresión lineal de orden > 1 sea más ajustado que uno de orden == 1.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Cargar dataset
df = sns.load_dataset('penguins').dropna()

# 2. Relación entre parejas de variables
sns.pairplot(df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']])
plt.show()

# 3. Correlación entre dos variables continuas
sns.scatterplot(data=df, x='bill_length_mm', y='flipper_length_mm')
plt.title('Correlación entre variables')
plt.show()

# 4. Dos variables numéricas según una categórica
sns.scatterplot(
    data=df,
    x='bill_length_mm',
    y='flipper_length_mm',
    hue='species'
)
plt.show()

# 5. Distribución de una variable
sns.boxplot(x=df['bill_length_mm'])
plt.title('Distribución y outliers')
plt.show()

# 6. Comparar densidad de dos variables
sns.kdeplot(df['bill_length_mm'], label='Bill length')
sns.kdeplot(df['flipper_length_mm'], label='Flipper length')
plt.legend()
plt.show()

# 7. Regresión no lineal
sns.regplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    order=2
)
plt.title('Regresión polinómica')
plt.show()
```
